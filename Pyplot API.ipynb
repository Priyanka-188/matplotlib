```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
interval_x = np.linspace(0,20,50)
test_data_y = interval_x
squared_test_data_y = interval_x**2
cubic_test_data_y = interval_x**3
quadratic_test_data = interval_x**4
#interval_x.shape
#interval_x.size
```

```python
plt.plot(interval_x,test_data_y)
plt.show()
```

```python
plt.plot(interval_x,squared_test_data_y )
plt.show()
```

```python
plt.plot(interval_x,cubic_test_data_y )
plt.show()
```

```python
plt.plot(interval_x,quadratic_test_data )
plt.show()
```

```python
plt.plot(interval_x,test_data_y,'orange',linewidth = 5)
plt.show()
```

# changing style of plot

```python
plt.plot(interval_x,squared_test_data_y ,'w',linewidth = 3)
print(plt.style.available)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [6,4]
```

```python
plt.plot(interval_x,test_data_y,'orange',linewidth = 5)
plt.style.use('seaborn-bright')
plt.show()
```
