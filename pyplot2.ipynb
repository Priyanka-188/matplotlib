```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,20,50)
y = x**2
z = x**3
```

```python
plt.plot(x,y,'r*',markersize = 7)
plt.show()
```

# line style

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'b',linewidth=3,linestyle = 'dotted')
plt.show()
```

```python
plt.plot(x,y,'g',linewidth = 3,linestyle = 'dashed')
plt.show()
```

```python
plt.plot(x,y,'g',linewidth = 3,linestyle = 'dashdot')
plt.show()
```

#  Marker type

```python
plt.plot(x,y,'g*',markersize = 7)
plt.show()
```

```python
plt.plot(x,y,'co',markersize = 7)
plt.show()
```

# multiplot

```python
plt.subplot(2,1,1)
plt.plot(x,y,'b--')
plt.xlabel('x')
plt.ylabel("square of x")

plt.subplots_adjust(hspace =0.5)
plt.subplot(2,1,2)
plt.plot(x,z,'r*')
plt.xlabel('x')
plt.ylabel("cube of x")
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'b--')
plt.xlabel('x')
plt.ylabel("square of x")


plt.subplots_adjust(wspace =1)
plt.subplot(2,2,2)
plt.plot(x,z,'r*')
plt.xlabel('x')
plt.ylabel("cube of x")
plt.show()
```

```python
plt.plot(x,y)
plt.text(2,380,'dummy text',fontsize = 20)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.5,0.5])
axes2 = fig.add_axes([0.7,0,1.2,1.2])

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('x square')
axes1.set_title("Square of number")

axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel("x square")
axes2.set_title('Square of no.2')

plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,0.9,0.9])
axes2 = fig.add_axes([0.15,0.5,0.3,0.3])

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('x square')
axes1.set_title("Square of number")

axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel("x square")
axes2.set_title('Square of no.2')

plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([1,0,1,1])
axes2 = fig.add_axes([0.5,0.5,0.3,0.3])

axes1.plot(x,y)
axes1.set_xlabel('x')
axes1.set_ylabel('x square')
axes1.set_title("Square of number")

axes2.plot(x,y)
axes2.set_xlabel('X')
axes2.set_ylabel("x square")
axes2.set_title('Square of no.2')

plt.show()
```
